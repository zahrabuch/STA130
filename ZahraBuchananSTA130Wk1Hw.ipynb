Q7.1 A use case in which .dropna() is preferred would be if someone was handling a data set on movies and they were looking to analyze the ratings of each movie but they end up finding a few missing values in the 'ratings' column. Instead of using del.movies['ratings'] they would use movies.dropna() to remove all rows with missing values for the rating. This is used instead as deleting the entire column is not an option because the data in that column is needed for analysis.

7.2 A use case for the opposite could be if someone was handling a movie data set and wanted to analyze the ratings of the movies but found the dataset had a column titled 'budget'(showing the budget for each movie) in which many of the values were missing. Since the person will not need the data in that column to analyze the ratings they could use del.movies['budget'] to remove the entire column.

7.3 It is important to use del before .dropna() as its more efficient and can prevent loss of important data. Since .dropna() removes all rows with missing values, if there is an unnecessary column with mainly missing values using .dropna() will remove all the corresponding rows and majority of your data will be gone. If del is used first to remove the unnecessary column then using .dropna() will be sure to only remove the rows not useful for analysis.
