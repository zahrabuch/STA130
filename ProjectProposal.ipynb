Research Question: Does relationship status affect the number of close friends a person may have?

Variables: Relationship Status(DEMO_relationship_status)This is a value that represents the relationship status associated with an observation. The options are In a relationship, single and dating and single but not dating. this is to ensure that each observation analyzed falls under a  relationship status and can be analysed accordingly.

Number of Close Friends(CONNECTION_social_num_close_friends_grouped)this is a range indicating the number of close friends held by someone. The options are: 1-2, 3-4 and 5 or more. This gives us the range of close friends for each observation which is needed for to answer the research question.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('CSCS_data_anon.csv',low_memory=False)
friends_col = "CONNECTION_social_num_close_friends_grouped"

data_friends = data[friends_col].dropna()
friends_counts = data_friends.value_counts().sort_index()

fig, ax = plt.subplots(figsize=(10, 6))
friends_counts.plot(kind='bar', color='skyblue', ax=ax)

ax.set_title("Distribution of Number of Close Friends")
ax.set_xlabel("Number of Close Friends")
ax.set_ylabel("Count")
fig.tight_layout()
fig.show()
fig.savefig("close_friends_distribution.png", format="png")
plt.show()

```

Research Question: Does gender influence the effectiveness of cat ownership in alleviating feelings of loneliness?

Variables: Gender(DEMO_gender)gives the gender of someone. This is necessary for grouping observations by gender.

Number of Cats Owned(GEO_housing_live_with_cats)this gives the number of cats owned by someone. This is needed in order to find all entries containing people who own cats. 

Loneliness(LONELY_ucla_loneliness_scale_companionship)this variable gives the degree to which one may lack companionship based on three anser choices: Hardly Ever, Some of the Time, and Often. This is needed in order to compare the degree of loneliness of the different gender group cat owners.

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('CSCS_data_anon.csv', low_memory=False)

gender_col = "DEMO_gender"

data_gender = data[gender_col].dropna()

gender_counts = data_gender.value_counts().sort_index()

fig, ax = plt.subplots(figsize=(10, 6))
gender_counts.plot(kind='bar', color='skyblue', ax=ax)

ax.set_title("Distribution of Gender")
ax.set_xlabel("Gender")
ax.set_ylabel("Count")
fig.savefig("gender_distribution.png", format="png")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('CSCS_data_anon.csv', low_memory=False)
housing_cats_col = "GEO_housing_live_with_cats"
data_housing_cats = data[housing_cats_col].dropna()
data_housing_cats = data_housing_cats[data_housing_cats != 0]
housing_cats_counts = data_housing_cats.value_counts().sort_index()

fig, ax = plt.subplots(figsize=(10, 6))
housing_cats_counts.plot(kind='bar', color='skyblue', ax=ax)

ax.set_title("Distribution of Living Situation with Cats")
ax.set_xlabel("Living Situation with Cats")
ax.set_ylabel("Count")

fig.savefig("housing_with_cats_distribution.png", format="png")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('CSCS_data_anon.csv', low_memory=False)

loneliness_col = "LONELY_ucla_loneliness_scale_companionship"
data_loneliness = data[loneliness_col].dropna()
loneliness_counts = data_loneliness.value_counts().sort_index()

fig, ax = plt.subplots(figsize=(10, 6))
loneliness_counts.plot(kind='bar', color='skyblue', ax=ax)

ax.set_title("Distribution of How often Persons Lack Companionship")
ax.set_xlabel("Loneliness Scores")
ax.set_ylabel("Count")
fig.savefig("loneliness_distribution.png", format="png")
plt.tight_layout()
plt.show()
```

Analysis(Method): I would use bootstrapping to obtain a confidence interval which infers the level of which owning a cat is effective in allieviating feelings of loneliness for each gender. The null hypothesis I would use is that owning a cat has no effect on allieviating feelings of loneliness and then try to disprove it for each gender and then compare the confidence intervals of each.

Hypothesis/Results: I hypothesize that for women, cats are more effective in allieviating feelings of loneliness than for men. I can also hypothesize that the evidence to reject the null hypothesis will be stronger for women than men.
